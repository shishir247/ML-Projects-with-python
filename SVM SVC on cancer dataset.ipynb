```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings(action = "ignore")
```

```python
data = pd.read_csv(r"F:\anaconda\data sets\breast-cancer-data.csv")
data.head()
```

```python
data.diagnosis = data["diagnosis"].replace("M" , 0)
data.diagnosis= data["diagnosis"].replace("B" , 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X = data.drop(["diagnosis", "id"], axis=1)
y = data.diagnosis

from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.2, random_state = 6)

xtrain_sc = sc.fit_transform(xtrain)
xtest_sc = sc.transform(xtest)

from sklearn.svm import SVC
svm = SVC()


svm.fit(xtrain_sc,ytrain)

print(svm.score(xtrain_sc, ytrain))
print(svm.score(xtest_sc, ytest))
```

```python
from sklearn.metrics import r2_score
score= r2_score(ytest,svm.predict(xtest_sc))
score
```
